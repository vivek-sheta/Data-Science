 
isna().any().any() return bool [See if there is any null value in our data-set]
isna().sum().sum() return total count how many missing values in data-set

```python
import pandas as pd
import numpy as np
df = pd.read_csv('property data.csv')
df.head()
df.isna().any()
#df.head()
```

Find out relateive missing values in dataset by columns 

```python
df.isna().sum()/len(df)*100
```

by chance if more then one coloum have missing value then get name of that columns

```python
df.loc[:,df.isnull().any()].columns
```

```python
df = df.replace(r'^--|^na', np.nan, regex=True)
df['PID'].fillna(df['PID'].mean(),inplace=True)
df['ST_NUM'].fillna('DUMMY',inplace=True)
df['OWN_OCCUPIED'].fillna('DUMMY',inplace=True)
df['NUM_BEDROOMS'].fillna('DUMMY',inplace=True)
df['NUM_BATH'].fillna('DUMMY',inplace=True)
df['SQ_FT'].fillna('DUMMY',inplace=True)
df
```

```python
df.isna().sum()/len(df)*100  #All missing values are filled
```

```python
pd.get_dummies(df.ST_NAME,prefix='ST_NAME')
```

```python
country_dummies = pd.get_dummies(df.ST_NAME,prefix='ST_NAME')
df = pd.concat([df,country_dummies],axis=1)
df
```

```python
df.drop(df.columns[[2]],axis=1,inplace=True)
df
```
